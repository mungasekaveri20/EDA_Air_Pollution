```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load your dataset
df = pd.read_csv('fds2.csv')
```

```python
df.shape
```

```python
df['AirQualityStatus'] = "Poor" if df['AirQuality'] > 50 else "Good"
```

```python
df.info()
```

```python
df.describe()
```

```python
# Histogram for AirQuality
plt.figure(figsize=(10, 6))
sns.histplot(df['AirQuality'], bins=20, kde=True, color='red')
plt.title('Distribution of AirQuality')
plt.xlabel('AirQuality')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['WaterPollution'], bins=20, kde=True, color='blue')
plt.title('Distribution of WaterPollution')
plt.xlabel('WaterPollution')
plt.ylabel('Frequency')
plt.show()
```

```python

# Calculate the correlation matrix
correlation_matrix = df[['AirQuality', 'WaterPollution']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap between AirQuality and WaterPollution')
plt.show()
```

```python
plt.figure(figsize=(12,6))
tmp_df = df.groupby("Country").agg({"AirQuality":"median"}).sort_values("AirQuality",ascending=False)[:50]
sns.barplot(x=tmp_df.index,y=tmp_df["AirQuality"])
plt.xticks(rotation=90)
plt.title("Top 50 countries with AirQuality")
plt.show()
```

```python
plt.figure(figsize=(12,6))
tmp_df = df.groupby("Country").agg({"WaterPollution":"median"}).sort_values("WaterPollution",ascending=False)[:50]
sns.barplot(x=tmp_df.index,y=tmp_df["WaterPollution"])
plt.xticks(rotation=90)
plt.title("Top 50 countries with WaterPollution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(y='AirQuality', x='Country', data=df, color='red', alpha=0.5, hue='Country')
plt.title('Scatter Plot of AirQuality vs. Country')
plt.ylabel('AirQuality')
plt.xlabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(y='WaterPollution', x='Country', data=df, color='red', alpha=0.5, hue='Country')
plt.title('Scatter Plot of Country vs. WaterPollution')
plt.ylabel('WaterPollution')
plt.xlabel('Country')
plt.show()
```
